```python
from matplotlib import pylab 
print(pylab.__version__)
```

```python
import numpy as np

X = np.linspace(0,10,25)
Y = X * X + 2
print(X)
print(Y)
print(np.array([X,Y]).reshape(25,2).reshape(2,25))
```

```python
pylab.plot(X,Y,'r')
```

```python
pylab.subplot(1,2,1)
pylab.plot(X,Y,'r--')
pylab.subplot(1,2,2)
pylab.plot(Y,X,'g*-')
```

```python
pylab.subplot(1,2,1)
pylab.plot(X,Y,'r--')
pylab.subplot(1,2,1)
pylab.plot(Y,X,'g*-')
```

```python
from matplotlib import pyplot as plt
```

```python
Fig = plt.figure()
Axis = Fig.add_axes([0.5,0.1,0.8,0.8])
Axis.plot(X,Y,'r')
```

```python
Fig, Axes = plt.subplots(nrows=1, ncols=2)
for ax in Axes:
    ax.plot(X,Y,'r')
```

```python
Fig = plt.figure()
Axes1 = Fig.add_axes([0.1,0.1,0.8,0.8])
Axes2 = Fig.add_axes([0.2,0.5,0.2,0.2])
Axes1.plot(X,Y,'r')
Axes2.plot(Y,X,'g')
```

```python
Fig = plt.figure(figsize=(16,9), dpi=100)
Axis = Fig.add_axes([0.5,0.1,0.8,0.8])
Axis.plot(X,Y,'r')
```
